```python
import sys
sys.path.append("../../..")
import fridom.NonHydrostatic as fr
```

```python
class MyPlotter(fr.ModelPlotter):
    def update_figure(fig, z, p, time, **kwargs):
        fr.Plot(z.b).front(z, fig=fig, cmax=1.2, vmax=1.9)
        fig.gca().set_title(f"Front View,   t = {time:.3f}")

mset = fr.ModelSettings(L=[3, 0.2, 3], N=[200, 5, 200])
mset.periodic_bounds[2] = False
mset.enable_nonlinear = False
mset.enable_vid_anim = True
mset.vid_plotter = MyPlotter
mset.vid_anim_interval = 10
mset.vid_anim_filename = "vertical_mode.mp4"

grid = fr.Grid(mset)
cp = grid.cp

z_ini = fr.InitialConditions.VerticalMode(
    mset, grid, kx=7, kz=1.5*2*cp.pi/3, phase=0, s=1)
dp = z_ini.kx * mset.dx
z_ini = fr.InitialConditions.VerticalMode(
    mset, grid, kx=7, kz=1.5*2*cp.pi/3, phase=dp, s=1)

model = fr.Model(mset, grid)
model.z = z_ini
model.run(runlen=z_ini.period)
model.show_video()
```
