```python
import sys
sys.path.append("../../..")

from fridom.NonHydrostatic.ModelSettings import ModelSettings
from fridom.NonHydrostatic.Grid import Grid
from fridom.NonHydrostatic.InitialConditions import VerticalMode
from fridom.NonHydrostatic.Model import Model
from fridom.NonHydrostatic.Plot import Plot
from fridom.NonHydrostatic.ModelPlotter import ModelPlotter
```

```python
class MyPlotter(ModelPlotter):
    def update_figure(fig, z, p, time, **kwargs):
        Plot(z.b).front(z, fig=fig, cmax=1.2, vmax=1.9)
        fig.gca().set_title(f"Front View,   t = {time:.3f}")

mset = ModelSettings(L=[3, 0.2, 3], N=[200, 5, 200])
mset.periodic_bounds[2] = False
mset.enable_nonlinear = False
mset.enable_vid_anim = True
mset.vid_plotter = MyPlotter
mset.vid_anim_interval = 10
mset.vid_anim_filename = "vertical_mode.mp4"

grid = Grid(mset)
cp = grid.cp

z_ini = VerticalMode(mset, grid, kx=7, kz=1.5*2*cp.pi/3, phase=0, s=1)
dp = z_ini.kx * mset.dx
z_ini = VerticalMode(mset, grid, kx=7, kz=1.5*2*cp.pi/3, phase=dp, s=1)

model = Model(mset, grid)
model.z = z_ini
model.run(runlen=z_ini.period)
model.show_video()
```
