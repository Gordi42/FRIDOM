```python
import sys
sys.path.append("../../..")
import fridom.NonHydrostatic as nh
```

```python
class MyPlotter(nh.ModelPlotter):
    def update_figure(fig, z, p, time, **kwargs):
        nh.Plot(z.b).front(fig=fig, cmax=0.1, vmax=0.1)
        fig.gca().set_title(f"Front View,   t = {time:.3f}")

mset = nh.ModelSettings(L=[3, 0.2, 3], N=[500, 2, 500], dt=0.01)
mset.enable_source = True
mset.enable_nonlinear = False
mset.periodic_bounds = [False, True, False]
mset.enable_vid_anim = True
mset.vid_plotter = MyPlotter
mset.vid_anim_interval = 25
mset.vid_anim_filename = "wave_ray.mp4"
grid = nh.Grid(mset)

wm = nh.Source.PolarizedWaveMaker(mset, grid, mask_pos=(0.5,None,0.5),
                        kx=80, kz=-40, amplitude=0.5, mask_width=(0.1,None,0.1))

model = nh.Model(mset, grid)
model.source_tendency.z = wm

model.run(runlen=20)
model.show_video()
```
