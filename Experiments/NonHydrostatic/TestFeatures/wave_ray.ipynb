```python
import sys
sys.path.append("../../..")

from FIDOM.NonHydrostatic.ModelSettings import ModelSettings
from FIDOM.NonHydrostatic.Grid import Grid
from FIDOM.NonHydrostatic.Model import Model
from FIDOM.NonHydrostatic.Plot import Plot
from FIDOM.NonHydrostatic.ModelPlotter import ModelPlotter
from FIDOM.NonHydrostatic.Source import PolarizedWaveMaker
```

```python
class MyPlotter(ModelPlotter):
    def update_figure(fig, z, p, time, **kwargs):
        Plot(z.b).front(fig=fig, cmax=1, vmax=1)
        fig.gca().set_title(f"Front View,   t = {time:.3f}")

mset = ModelSettings(L=[3, 0.2, 3], N=[500, 2, 500], dt=0.01)
mset.enable_source = True
mset.enable_nonlinear = False
mset.periodic_bounds = [False, True, False]
mset.enable_vid_anim = True
mset.vid_plotter = MyPlotter
mset.vid_anim_interval = 25
mset.vid_anim_filename = "wave_ray.mp4"
grid = Grid(mset)

wm = PolarizedWaveMaker(mset, grid, mask_pos=(0.5,None,0.5),
                        kx=80, kz=-40, amplitude=0.5, mask_width=(0.1,None,0.1))

model = Model(mset, grid)
model.source = wm

model.run(runlen=100)
model.show_video()
```
